```python
# import the stuff needed
import mesa_reader as mr
from math import log
import matplotlib.pylab as plt
```

```python
# get the data into useable format
g = mr.MesaData("history2.data")

masses = g.star_mass # units in Msun
logMasses = [log(y,10) for y in masses] # make it all log10 for consistency

radii = g.log_R # units in Rsun

age = g.star_age # units in years

coreHe = g.he_core_mass
coreC = g.c_core_mass
coreMass = coreHe + coreC # units probably in Msun but MESA doesn't specify
```

```python
# plot the radius 
plt.plot(age, radii)
plt.xlabel('Years')
plt.ylabel('log Radius (Rsun)')
plt.grid(axis='both')
plt.show()
```

```python
# plot the mass
plt.plot(age, logMasses)
plt.xlabel('Years')
plt.ylabel('log Mass (Msun)')
plt.show()
```

```python
# plot the core masses
plt.plot(age, coreMass, linestyle='solid')
plt.plot(age, coreHe, linestyle='--')
plt.plot(age, coreC, linestyle=':')
plt.xlabel('log Years')
plt.ylabel('He, C, and Total Core Mass (Msun)')
plt.xscale('log')
plt.grid(axis='both')
plt.show()
```
