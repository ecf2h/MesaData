```python
# import the stuff needed
import mesa_reader as mr
from math import log
import matplotlib.pylab as plt

# define the temperature limits here
maxTemp = 4.7
minTemp = 3.5
```

```python
# frankenstein the data together because MESA is annoying
hBR = mr.MesaData('historyBeforeRemove.data')
hR = mr.MesaData('historyRemove.data')
hAR = mr.MesaData('historyAfterRemove.data')
htSB = mr.MesaData('historyToSiBurn.data')
htlgT = mr.MesaData('historyTolgT.data')
hC = mr.MesaData('historyConvert.data')
hF = mr.MesaData('historyFinish.data')

lumBR = hBR.log_L
effTempBR = hBR.log_Teff
ageBR = hBR.star_age
rBR = hBR.log_R

lumR = hR.log_L
effTempR = hR.log_Teff
ageR = hR.star_age
rR = hR.log_R

lumAR = hAR.log_L
effTempAR = hAR.log_Teff
ageAR = hAR.star_age
rAR = hAR.log_R

lumtSB = htSB.log_L
effTemptSB = htSB.log_Teff
agetSB = htSB.star_age
rtSB = htSB.log_R

lumtlgT = htlgT.log_L
effTemptlgT = htlgT.log_Teff
agetlgT = htlgT.star_age
rtlgT = htlgT.log_R

lumC = hC.log_L
effTempC = hC.log_Teff
ageC = hC.star_age
rC = hC.log_R

lumF = hF.log_L
effTempF = hF.log_Teff
ageF = hF.star_age
rF = hF.log_R
```

```python
# make a HR diagram - plot
plt.figure(figsize=(15,4))
plt.plot(effTempBR, lumBR)
plt.plot(effTempR, lumR)
plt.plot(effTempAR, lumAR)
plt.plot(effTemptSB, lumtSB)
plt.plot(effTemptlgT, lumtlgT)
plt.plot(effTempC, lumC)
#plt.plot(effTempF, lumF)
plt.xlabel('log Effective Temperature (reversed)')
plt.ylabel('log Luminosity (Lsun)')
plt.grid(axis='both')
plt.xlim(maxTemp, minTemp)
#plt.plot(effTemp[0], lum[0], 'ro')
#plt.plot(effTemp[last], lum[last], 'r*')
plt.show()
```

```python
# make a HR diagram - scatter
plt.figure(figsize=(15,4))
plt.scatter(effTempBR, lumBR)
plt.scatter(effTempR, lumR)
plt.scatter(effTempAR, lumAR)
plt.scatter(effTemptSB, lumtSB)
plt.scatter(effTemptlgT, lumtlgT)
plt.scatter(effTempC, lumC)
#plt.scatter(effTempF, lumF)
plt.xlabel('log Effective Temperature (reversed)')
plt.ylabel('log Luminosity (Lsun)')
plt.grid(axis='both')
plt.xlim(maxTemp, minTemp)
```

```python
# plot the radius over time
plt.figure(figsize=(15,4))
plt.plot(ageBR, rBR)
plt.plot(ageR, rR)
plt.plot(ageAR, rAR)
plt.plot(agetSB, rtSB)
plt.plot(agetlgT, rtlgT)
plt.plot(ageC, rC)
#plt.plot(ageF, rF)
plt.xlabel('Age in Years')
plt.ylabel('log Radius (Rsun)')
plt.grid(axis='both')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
```
